```python
import seaborn as sns
import pandas as pd
```

```python
df = pd.DataFrame(data.data,columns=data.feature_names)
```

```python
df['Species'] = data.target
```

```python
df.loc[df['Species']==0,"Species"] = data.target_names[0]
df.loc[df['Species']==1,"Species"] = data.target_names[1]
df.loc[df['Species']==2,"Species"] = data.target_names[2]
df
```

```python
sns.pairplot(df, hue="Species")
plt.show()
```

```python
from pandas.plotting import parallel_coordinates
parallel_coordinates(df, "Species",colormap='cool')
plt.show()
```
