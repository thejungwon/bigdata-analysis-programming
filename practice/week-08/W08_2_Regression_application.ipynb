## 2. Feature 선택

```python
selected_columns = ["LotArea","OverallQual","OverallCond","YearBuilt", "YearRemodAdd",
                   "Heating","HeatingQC","CentralAir","KitchenQual"]
new_df = df[selected_columns].copy()

new_df.head()
```

## 3. Feature 인코딩

```python
new_df.loc[new_df['Heating']=="GasA",'Heating'] = 0
new_df.loc[new_df['Heating']=="GasW",'Heating'] = 1
new_df.loc[new_df['Heating']=="Grav",'Heating'] = 2
new_df.loc[new_df['Heating']=="Wall",'Heating'] = 3
new_df.loc[new_df['Heating']=="OthW",'Heating'] = 4
new_df.loc[new_df['Heating']=="Floor",'Heating'] = 5
```

```python
new_df.loc[new_df['HeatingQC']=="Ex",'HeatingQC'] = 0
new_df.loc[new_df['HeatingQC']=="TA",'HeatingQC'] = 1
new_df.loc[new_df['HeatingQC']=="Gd",'HeatingQC'] = 2
new_df.loc[new_df['HeatingQC']=="Fa",'HeatingQC'] = 3
new_df.loc[new_df['HeatingQC']=="Po",'HeatingQC'] = 4
```

```python
new_df.loc[new_df['KitchenQual']=="Ex",'KitchenQual'] = 0
new_df.loc[new_df['KitchenQual']=="TA",'KitchenQual'] = 1
new_df.loc[new_df['KitchenQual']=="Gd",'KitchenQual'] = 2
new_df.loc[new_df['KitchenQual']=="Fa",'KitchenQual'] = 3
```

```python
new_df.loc[new_df['CentralAir']=="Y",'CentralAir'] = 0
new_df.loc[new_df['CentralAir']=="N",'CentralAir'] = 1
```

```python
new_df.head()
```

```python
# Numpy 포맷으로 변경
X = new_df.to_numpy()
y = df['SalePrice'].to_numpy()
```

```python
from sklearn.model_selection import train_test_split
#훈련/테스트 데이터 셋 분리
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
import numpy as np
from sklearn.linear_model import LinearRegression
#학습
reg = LinearRegression().fit(X_train, y_train)
#예측
y_pred = reg.predict(X_test)
```

```python
#예측 결과
y_pred
```

```python
#정답
y_test
```

## 4. 모델 검증

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
from math import sqrt


print("MSE: ",mean_squared_error(y_test, y_pred))
print("MAE: ",mean_absolute_error(y_test, y_pred))
print("RMSE: ",sqrt(mean_squared_error(y_test, y_pred)))
```

```python
from sklearn.svm import SVR
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import AdaBoostRegressor
from sklearn.neural_network import MLPRegressor
from sklearn import linear_model
regressors = [SVR(), DecisionTreeRegressor(),RandomForestRegressor(), AdaBoostRegressor(),
              MLPRegressor(max_iter=5000), linear_model.Lasso(alpha=0.1)]
best_model = {}
best_score = float('inf')
for reg in regressors:
    reg.fit(X_train, y_train)
    y_pred = reg.predict(X_test)

    name = reg.__class__.__name__
    mse = mean_squared_error(y_test, y_pred)
    mae = mean_absolute_error(y_test, y_pred)
    rmse = sqrt(mean_squared_error(y_test, y_pred))
    print("Name:",name)
    print("MSE: ",mse)
    print("MAE: ",mae)
    print("RMSE: ",rmse)
    
    if mae<best_score:
        best_score = mae
        best_model['model'] = reg
        best_model['mae'] = mae
        best_model['mse'] = mse
        best_model['rmse'] = rmse
    
    print("="*30)
```
